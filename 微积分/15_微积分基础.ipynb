# 极限

## 常用函数的极限I

```python
from sympy import *

x = symbols('x')
k = symbols('k')
a = symbols('a')
b = symbols('b')
c = symbols('c')
n = symbols('n')

functions = [
    k,
    x,
    b*x+c,
    x**n,
]

for f in functions:
    print(f"lim({f})，x趋近a时，极限为 {limit(f, x, a)}")
```

```python
from sympy import *

x = symbols('x')
n = symbols('n')
f = x**n

lmt = limit(f.subs(n, 2), x, 0)
print(f"lim({f})，n>0，x趋近0时，极限为 {lmt}")

lmt = limit(f.subs(n, -2), x, oo)
print(f"lim({f})，n<0，x趋近oo时，极限为 {lmt}")

lmt = limit(f.subs(n, -2), x, -oo)
print(f"lim({f})，n<0，x趋近-oo时，极限为 {lmt}")

lmt = limit(f.subs(n, -2), x, 0, '+')
print(f"lim({f})，n<0，x从右边趋近0时，极限为 {lmt}")

lmt = limit(f.subs(n, -2), x, 0, '-')
print(f"lim({f})，n<0且是偶数，x从左边趋近0时，极限为 {lmt}")

lmt = limit(f.subs(n, -3), x, 0, '-')
print(f"lim({f})，n<0且是奇数，x从左边趋近0时，极限为 {lmt}")
```

## 极限的算数运算

```python
from sympy import *

x = symbols('x')
k = symbols('k')
a = symbols('a')
b = symbols('b')
c = symbols('c')
n = symbols('n')

f = 1/x**2
g = sin(x)

print(f"lim({f}+{g})，x趋近a时，极限为 {limit(f+g, x, a)}")
print(f"lim({f})+lim({g})，x趋近a时，极限为 {limit(f, x, a)} + {limit(g, x, a)}")
print(f'相等：{(limit(f, x, a) + limit(g, x, a)).equals(limit(f+g, x, a))}')

print('')
print(f"lim({f}-{g})，x趋近a时，极限为 {limit(f-g, x, a)}")
print(f"lim({f})-lim({g})，x趋近a时，极限为 {limit(f, x, a)} - {limit(g, x, a)}")
print(f'相等：{(limit(f, x, a) - limit(g, x, a)).equals(limit(f-g, x, a))}')

print('')
print(f"lim({f}*{g})，x趋近a时，极限为 {limit(f*g, x, a)}")
print(f"lim({f})*lim({g})，x趋近a时，极限为 {limit(f, x, a)} * {limit(g, x, a)}")
print(f'相等：{(limit(f, x, a) * limit(g, x, a)).equals(limit(f*g, x, a))}')

print('')
print(f"lim({f}/{g})，x趋近a时，极限为 {limit(f/g, x, a)}")
print(f"lim({f})/lim({g})，x趋近a时，极限为 {limit(f, x, a)} / {limit(g, x, a)}")
print(f'相等：{(limit(f, x, a) / limit(g, x, a)).equals(limit(f/g, x, a))}')
```

## 常见函数的极限II

```python
from sympy import *

x = symbols('x')
k = symbols('k')
a = symbols('a')
b = symbols('b')
c = symbols('c')
n = symbols('n')

functions = [
    sin(x)/x,
    tan(x)/x,
    (1-cos(x))/x,
    cos(x),
    E**x,
    (E**x-1)/x,
]

for f in functions:
    print(f"lim({simplify(f)})，x趋近0时，极限为 {simplify(limit(f, x, 0))}")

f1 = (x**n-a**n)/(x-a)
print(f"lim({f1})，x趋近a时，极限为 {limit(f1.subs(n, 4), x, a)}")

f2 = (1+1/x)**x
print(f"lim({simplify(f2)})，x趋近oo时，极限为 {limit(f2, x, oo)}")
```

## 有理化求极限

$\lim _{x\rightarrow 2}\dfrac{x-2}{x^{2}-4}=\lim _{x\rightarrow 2}\dfrac{x-2}{\left( x+2\right) \left( x-2\right)}=\lim _{x\rightarrow 2}\dfrac{1}{x+2} = \dfrac{1}{4}$

```python
from sympy import *

x = symbols('x')

f = (x-2)/(x**2-4)
print(limit(f, x, 2))
```

# 导数

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation

plt.rcParams["animation.html"] = "jshtml"
plt.ioff()
fig, ax = plt.subplots()

x = sp.symbols('x')
f = x**2 + 2*x + 1
df = sp.diff(f, x)

x_point = 1
y_point = f.subs(x, x_point)
slope_at_point = df.subs(x, x_point)

start = 2
end = 0.1
frames = 11
step = (end-start) / (frames-1)


def animate(t):
    plt.cla()

    deltaX = start + step * t
    x_point2 = x_point + deltaX
    y_point2 = f.subs(x, x_point2)
    slope2 = (y_point2-y_point)/(x_point2-x_point)

    x_vals = np.linspace(-5, 5, 400)
    y_vals = [f.subs(x, val) for val in x_vals]
    plt.plot(x_vals, y_vals, label=f'$f(x) = {sp.latex(f)}$')

    y_intercept = y_point - slope2 * x_point
    x_line = np.linspace(x_point-1, x_point2+1, 10)
    y_line = slope2 * x_line + y_intercept
    plt.plot(x_line, y_line, '--g', label='line of 2 points')

    x_line = np.linspace(x_point - 1, x_point + 1, 10)
    y_line = [slope_at_point * (val - x_point) + y_point for val in x_line]
    plt.plot(x_line, y_line, 'r--', label='derivative line')

    plt.scatter([x_point, x_point2], [y_point, y_point2], color='g')

    plt.ylim(-1, 40)
    plt.legend()
    plt.grid(True)


matplotlib.animation.FuncAnimation(fig, animate, frames=frames)
```

## 符号
