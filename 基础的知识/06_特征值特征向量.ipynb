# 分解特征值特征向量

```python
import numpy as np

A = np.arange(1, 10).reshape(3,3)
print('A:\n', A)

values, vectors = np.linalg.eig(A)
print('values:\n', values)
print('vectors:\n', vectors)
```

# 验证特征值特征向量

```python
import numpy as np


A = np.arange(1, 10).reshape(3,3)
print('A:\n', A)

values, vectors = np.linalg.eig(A)

B = A.dot(vectors[:, 0])
print('B:\n', B)

C = vectors[:, 0] * values[0]
print('C:\n', C)
```

# 重构原始矩阵

```python
import numpy as np

A = np.arange(1, 10).reshape(3,3)
print('A:\n', A)

values, vectors = np.linalg.eig(A)


Q = vectors
R = np.linalg.inv(Q)
L = np.diag(values)

B = Q.dot(L).dot(R)
print(B)
```

# 空间线性变换中的特征向量

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建一个网格
X, Y = np.mgrid[-10:10:20j, -10:10:20j]

# 收敛
A = np.array([
    [1, 1],
    [1, 0],
])
U = (X + Y) * 5
V = X * 5


# # 发散
# A = np.array([
#     [3, -1],
#     [0, 1],
# ])
# U = (X * 3 - Y) * 5
# V = Y * 5

# # 旋转
# U = (-Y) * 5
# V = X * 5

values, vectors = np.linalg.eig(A)
print('values: ', values)
print('vectors: \n', vectors)

norm = np.sqrt(U**2 + V**2)
U_norm = U * 40 / norm
V_norm = V * 40 / norm




# 绘制矢量场
fig, ax = plt.subplots(figsize=(8, 8))
q = ax.quiver(X, Y, U_norm, V_norm, units='xy', scale=60, angles='uv', pivot='mid', color='blue')

v1 = values[0] * vectors[:, 0] * 200
v2 = values[1] * vectors[:, 1] * 200

ax.plot([0, v1[0]], [0, v1[1]])
ax.plot([0, v2[0]], [0, v2[1]])

# 画数列
a = [0, 1]
for i in range(10):
    ax.plot(a[0], a[1], 'o', color='red')
    a = A.dot(a)

    
    
# 设置坐标轴范围并显示网格
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.grid(True)
ax.set_aspect('equal', adjustable='box')

plt.show()
```
