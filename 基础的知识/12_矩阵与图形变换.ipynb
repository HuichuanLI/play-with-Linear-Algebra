# 缩放 - 标量

```python
import matplotlib.animation
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["animation.html"] = "jshtml"
plt.ioff()
fig, ax = plt.subplots()

points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1],
    [0, 0],
])

start = 1
end = 3
frames = 11
step = (end-start) / (frames-1)

def animate(t):
    plt.cla()
    T = start + step * t

    transformed_points = T * points

    plt.xlim(-0.1, 5)
    plt.ylim(-0.1, 5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.plot(transformed_points[:, 0], transformed_points[:, 1], 'ro-')


matplotlib.animation.FuncAnimation(fig, animate, frames=frames)
```

# 旋转

```python
import matplotlib.animation
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["animation.html"] = "jshtml"
plt.ioff()
fig, ax = plt.subplots()


points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1],
    [0, 0],
])


start = 0
end = 360
frames = 25
step = (end-start) / (frames-1)

def animate(t):
    radians = np.deg2rad(start + step*t)
    
    plt.cla()
    T = np.array([
        [np.cos(radians), -np.sin(radians)],
        [np.sin(radians), np.cos(radians)],
    ])
    transformed_points = np.dot(T, points.T).T

    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.plot(transformed_points[:, 0], transformed_points[:, 1], 'ro-')
    plt.grid(True)


matplotlib.animation.FuncAnimation(fig, animate, frames=frames)
```

# 齐次坐标 - 移动

```python
import matplotlib.animation
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["animation.html"] = "jshtml"
plt.ioff()
fig, ax = plt.subplots()


points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1],
    [0, 0],
])
points = np.column_stack((points, np.ones(points.shape[0])))

start = 0
end = 2
frames = 11
step = (end-start) / (frames-1)

def animate(t):

    plt.cla()

    T = np.array([
        [1, 0, start+t*step],
        [0, 1, start+t*step],
        [0, 0, 1],
    ])
    transformed_points = np.dot(T, points.T).T

    plt.xlim(0, 4)
    plt.ylim(-2, 2)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.plot(transformed_points[:, 0], transformed_points[:, 1], 'ro-')
    plt.grid(True)


matplotlib.animation.FuncAnimation(fig, animate, frames=frames)
```

# 仿射变换 - 组合

```python
import matplotlib.animation
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["animation.html"] = "jshtml"
plt.ioff()
fig, ax = plt.subplots()


def affineTransform(points, sx=1, sy=1, rotateDeg=0, shx=0, shy=0, tx=0, ty=0):
    # A = np.diag(np.ones(3))

    c, s = np.cos(np.deg2rad(rotateDeg)), np.sin(np.deg2rad(rotateDeg))
    Ts = np.array([
        [sx, 0, 0],
        [0, sy, 0],
        [0, 0, 1]
    ])
    Tr = np.array([
        [c, -s, 0],
        [s, c, 0],
        [0, 0, 1]
    ])
    Tsh = np.array([
        [1, shx, 0],
        [shy, 1, 0],
        [0, 0, 1]
    ])
    Tt = np.array([
        [1, 0, tx],
        [0, 1, ty],
        [0, 0, 1]
    ])
    T = Tt.dot(Tsh).dot(Tr).dot(Ts)
    return (T.dot(points.T).T, T)


points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1],
    [0, 0],
])
points = np.column_stack((points, np.ones(points.shape[0])))


start = 0
end = 2
frames = 11
step = (end-start) / (frames-1)


def animate(t):
    plt.cla()

    transformed_points, T = affineTransform(points, sx=2, rotateDeg=45, shx=t/(frames-1))
    # transformed_points = affineTransform(points, sx=2, rotateDeg=30, shx=0, ty=2*t/(frames-1))
    # transformed_points, T = affineTransform(points, sx=2, rotateDeg=30, shx=t/(frames-1), ty=1)
    Tinv = np.linalg.inv(T)
    transformed_points = Tinv.dot(transformed_points.T).T

    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.plot(transformed_points[:, 0], transformed_points[:, 1], 'ro-')
    # plt.plot(testPoints[:, 0], testPoints[:, 1], 'ro-')

    plt.grid(True)


matplotlib.animation.FuncAnimation(fig, animate, frames=frames)
```

# 倾斜

```python
import matplotlib.animation
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["animation.html"] = "jshtml"
plt.rcParams['figure.dpi'] = 150
plt.ioff()
fig, ax = plt.subplots()


points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1],
    [0, 0],
])
points = np.column_stack((points, np.ones(points.shape[0])))


start = 0
end = 2
frames = 11
step = (end-start) / (frames-1)

def animate(t):
    radians = np.deg2rad(start + step*t)

    plt.cla()

    T = np.array([
        [1, 0, 0],
        [0, 1, 0],
        [start+t*step, 0, 1],
    ])
    transformed_points = np.dot(T, points.T).T

    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.plot(transformed_points[:, 0]/transformed_points[:, 2], transformed_points[:, 1]/transformed_points[:, 2], 'ro-')
    plt.grid(True)


matplotlib.animation.FuncAnimation(fig, animate, frames=frames)
```
