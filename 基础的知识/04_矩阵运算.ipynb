坐标变换

```python
import matplotlib.pyplot as plt

T = [[2,1],[1,2]]
A = [[1,1], [2,1], [2,2], [1,2]]

A1 = np.zeros((2, 5))
A1[:, :-1] = np.array(A).T
A1[:, -1] = np.array(A).T[:,0]
print(A1)

B = np.dot(T, np.array(A).T)
print(B)

B1 = np.zeros((2, 5))
B1[:, :-1] = B
B1[:, -1] = B[:,0]
print(B1)

f, ax = plt.subplots(1)
ax.plot(A1[0], A1[1], marker='o')
ax.plot(B1[0], B1[1], marker='o')

ax.set_xlim(xmin=0, xmax=10)
ax.set_ylim(ymin=0, ymax=10)
ax.set_aspect('equal')
plt.grid(True)

plt.show()
```

椭圆体

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# x = np.linspace(-2, 2, 100)
# y = np.linspace(-2, 2, 100)

x = np.linspace(0, 2 * np.pi, 100)
y = np.linspace(0, np.pi, 100)


X, Y = np.meshgrid(x, y)


fig = plt.figure(figsize=(10, 10))

def plot(ax, X, Y, elev, azim):


    x1 = 1 * np.cos(X) * np.sin(Y)
    y1 = 2 * np.sin(X) * np.sin(Y)
    z1 = 3 * np.cos(Y)
    # z1 = np.sqrt(9*(1 - x1**2 - 0.25*y1**2))
    ax.plot_surface(x1, y1, z1, cmap='viridis', edgecolor='none')


    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_box_aspect((1, 1, 1))

    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 3)
    ax.set_zlim(-3, 3)

    ax.view_init(elev=elev, azim=azim)


ax1 = fig.add_subplot(221, projection='3d')
ax2 = fig.add_subplot(222, projection='3d')
ax3 = fig.add_subplot(223, projection='3d')
ax4 = fig.add_subplot(224, projection='3d')
plot(ax1, X, Y, 5, 10)
plot(ax2, X, Y, 20, 30)
plot(ax3, X, Y, 30, 60)
plot(ax4, X, Y, 0, 85)

# 显示图形
plt.show()
```

用行列式求椭圆体体积

```python
a, b, c = 1, 2, 3

vol1 = (4/3) * np.pi * a * b * c

volSphere = (4/3) * np.pi
det = np.linalg.det(np.diag([a, b, c]))
vol2 = volSphere * det

print(f"体积1: {vol1:.2f}")

print(f"单位球体积: {volSphere:.2f}, det: {det}, 体积2: {vol2:.2f}")
```

行列式解线性方程组

```python
A = [[ 2, -3, 18],
	 [ 3,  1,  5],
	 [-3,  9,  7]]

B = [7, 9, 6]

D = np.linalg.det(A)

Dx = np.linalg.det([
	[7, -3, 18 ],
	[9,  1,  5 ],
	[6,  9,  7 ]])

Dy = np.linalg.det([
	[ 2,  7, 18],
	[ 3,  9,  5],
	[-3,  6,  7]])

Dz = np.linalg.det([
	[ 2, -3,  7],
	[ 3,  1,  9],
	[-3,  9,  6]])

x = Dx/D
y = Dy/D
z = Dz/D

print(x, y, z)

np.allclose(np.array(A).dot([x, y, z]), B)
```

稀疏矩阵

```python
from numpy import array
from scipy.sparse import csr_matrix

A = array([
[1, 0, 0, 1, 0, 0],
[0, 0, 2, 0, 0, 1],
[0, 0, 0, 2, 0, 0]])
print(A)

S = csr_matrix(A)
print(S)


B = S.todense()
print(B)
```

计算数组的稀疏度

```python
from numpy import array
from numpy import count_nonzero

A = array([
[1, 0, 0, 1, 0, 0],
[0, 0, 2, 0, 0, 1],
[0, 0, 0, 2, 0, 0]])
print(A)

sparsity = 1.0 - count_nonzero(A) / A.size
print(sparsity)
```
