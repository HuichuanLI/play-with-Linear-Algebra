```python
import matplotlib.pyplot as plt
```

```python
 points = [[0, 0], [0, 5], [3, 5], [3, 4], [1, 4],
              [1, 3], [2, 3], [2, 2], [1, 2], [1, 0]]
```

```python
x = [point[0] for point in points]
y = [point[1] for point in points]
```

```python
plt.figure(figsize=(5, 5))
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.plot(x, y)
```

```python
import numpy as np
```

```python
P = np.array(points)
```

```python
 Tran = np.array([[2, 0], [0, 1.5]])
```

```python
Tran.dot(P.T)
```

```python
plt.plot(Tran.dot(P.T)[0,],Tran.dot(P.T)[1,])
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.plot(x, y)
```

```python
Tran = np.array([[1, 0], [0, -1]])
plt.plot(Tran.dot(P.T)[0,],Tran.dot(P.T)[1,])
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.plot(x, y)
```

```python
Tran = np.array([[-1, 0], [0, 1]])
plt.plot(Tran.dot(P.T)[0,],Tran.dot(P.T)[1,])
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.plot(x, y)
```

```python
Tran = np.array([[-1, 0], [0, -1]])
plt.plot(Tran.dot(P.T)[0,],Tran.dot(P.T)[1,])
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.plot(x, y)
```

```python
Tran = np.array([[1, 0], [0.5, 1]])
plt.plot(Tran.dot(P.T)[0,],Tran.dot(P.T)[1,])
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.plot(x, y)
```

```python
import math 
theta = math.pi /3 
Tran = np.array([[math.cos(theta), math.sin(theta)], [-math.sin(theta), math.cos(theta)]])
plt.plot(Tran.dot(P.T)[0,],Tran.dot(P.T)[1,])
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.plot(x, y)
```
