绘制一个三维曲面

```python
# 创建一个网格
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# 定义曲面函数
r = np.sqrt(x**2 + y**2)
z = np.sin(r) / r

# 创建一个3D轴
fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')
ax1 = fig.add_subplot(122, projection='3d')

# 绘制曲面
ax.plot_surface(x, y, z, cmap='viridis', linewidth=0)
ax1.plot_surface(x, y, z)


# 添加坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 显示图形
plt.show()
```
