# 向量

定义向量

```python
v1 = np.array([1,2,3])
v2 = np.array([1,2,3], dtype=float)
v3 = np.array([1.,2,3])

print(v1.dtype, v1.shape, v1)
print(v2.dtype, v3.shape, v2)
print(v3.dtype, v3.shape, v3)

A = np.array([[1,2], [3,4]])
print(A.dtype, A.shape, A)
```

向量的加减

```python
a = np.array([2,3,5])
b = np.array([1,2,3])
c = a + b
print('a+b: ', c)
print('a-b: ', a - b)
```

向量的哈达玛积（Hadamard product）、除

```python
a = np.array([2,3,5])
b = np.array([1,2,3])

print('a∘b: ', a * b)
print('a/b: ', a / b)
print('标量向量乘: ', 3 * a, a * 3)
```

点积

```python
a = np.array([1, 2, 3])
b = np.array([2, 3, 4])
c = a.dot(b)

print('a∙b: ', c, np.dot(b, a))

print(np.dot(c, (a + b)))
print(np.dot(c, a) + np.dot(c, b))
```

叉积

```python
a = np.array([1, 2, 3])
b = np.array([2, 3, 4])
c = np.cross(a, b)
print('a×b: ', c)
```

```python
# import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure()

def plot(rci, a, b, c):

    ax = fig.add_subplot(rci, projection='3d')
    ax.quiver(0, 0, 0, a[0], a[1], a[2], color='r', label='Vector a')
    ax.quiver(0, 0, 0, b[0], b[1], b[2], color='b', label='Vector b')
    ax.quiver(0, 0, 0, c[0], c[1], c[2], color='g', label='Vector c (Cross Product)')

    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")

    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.set_zlim(0, 2)


a = np.array([1, 0, 0])
b = np.array([0, 1, 0])
c = np.cross(a, b)
plot(221, a, b, c)

a = np.array([1, 0, 0])
b = np.array([0.707, 0.707, 0])
c = np.cross(a, b)
plot(222, a, b, c)

a = np.array([1, 0, 0])
b = np.array([-0.707, 0.707, 0])
c = np.cross(a, b)
plot(223, a, b, c)

a = np.array([1, 0, 0])
b = np.array([2, 2, 0])
c = np.cross(a, b)
plot(224, a, b, c)


plt.show()
```

单位向量

```python
x = np.random.rand(10)
norm = np.linalg.norm(x)
unitx = x / norm

print(x)
print(norm)
print(unitx)

print(np.dot(unitx, x), np.dot(x, unitx))
```

向量的范数

```python
print(np.linalg.norm([1,2], 1))
print(np.linalg.norm([1,2], 2))
print(np.linalg.norm([1,-2], np.inf))
```

```python
# https://www.youtube.com/watch?v=It2g7sDxdqI
import numpy as np
import matplotlib.pyplot as plt

def plotNorm(ax, n, p):
    X = (2 * (np.random.rand(2, n)-0.5))
    ind = []
    for i in range(n):
        if np.linalg.norm(X[:, i], p) <= 1:
            ind.append(i)
    # plt.fill(X[0, ind], X[1, ind], 'go')
    # ax.plot(X[0, ind], X[1, ind], marker='o')

    ax.scatter(X[0, ind], X[1, ind], marker='o')

    ax.set_aspect('equal', 'box')
    ax.grid()


fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

plotNorm(ax1, 5000, 1)
plotNorm(ax2, 5000, 2)
plotNorm(ax3, 5000, 4)
plotNorm(ax4, 5000, np.inf)


# plt.grid()
plt.show()
```

# 矩阵

矩阵的基本运算

```python
A = np.array([[1,2], [3,4]])
B = np.array([[5,6], [7,8]])
print('A+B: \n', A+B)
print('A-B: \n', A-B)
print('A*B: \n', A*B)
print('A/B: \n', A/B)
```

矩阵的点积（相乘）

```python
A = np.array([[1,2], [3,4], [5,6]])
B = np.array([[5,6,7], [7,8,9]])

print(np.dot(A, B))
print(A.dot(B))
print(A @ B)
```

零矩阵

```python
np.zeros((3,2))
```

对角矩阵

```python
a = [1, 2, 3, 4, 4]
A = np.diag(a)
print('对角矩阵\n', A)
print('对角矩阵的对角向量', np.diag(A))

# 这种也是对角矩阵
B = np.array([
    [1,0,0,0], 
    [0,2,0,0], 
    [0,0,3,0], 
    [0,0,0,4], 
    [0,0,0,0]])

print('对角矩阵的对角向量2', np.diag(B))
```

单位矩阵

```python
np.identity(5)
```

标量矩阵

```python
k = 5
D = np.diag([k] * 3)
print(D)
print(3 * np.identity(5))
```

三角矩阵

```python
v = np.array([1, 2, 3])
# 第1个维度上重复3遍，第二个维度上重复1遍
A = np.tile(v, (3, 1))
print(A)
# print(np.tile(v, (3, 2)))

print('下三角矩阵\n', np.tril(A))
print('上三角矩阵\n', np.triu(A))
```

对称矩阵

```python
M = np.array([
    [1, 2, 3, 4],
    [2, 5, 6, 7],
    [3, 6, 8, 9],
    [4, 7, 9, 10]])

print(M)
print(M.T)
print('equals:', np.array_equal(M, M.T))
```

正交矩阵

```python
from numpy.linalg import inv

Q = np.array([
    [1, 0],
    [0, -1]])
print('Q:\n', Q)

QT = Q.T
print('QT:\n', Q.T)

V = inv(Q)
print('inv:\n', V)

print('QT==inv:', np.array_equal(QT, V))

I = Q.dot(QT)
print('Q∙QT: \n', I)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 创建三维坐标轴
fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')
ax1 = fig.add_subplot(122, projection='3d')


# 定义线的起点和终点
start = np.array([0, 0, 0])
end = np.array([1, 1, 1])

# 绘制线
ax.plot([start[0], end[0]], [start[1], end[1]], [start[2], end[2]], color='red')

ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_zlim(0, 5)

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.view_init(elev=10.)

# 绘制线
ax1.plot([start[0], end[0]], [start[1], end[1]], [start[2], end[2]], color='red')

ax1.set_xlim(0, 5)
ax1.set_ylim(0, 5)
ax1.set_zlim(0, 5)

# 设置坐标轴标签
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

ax1.view_init(elev=50.)


# 显示图形
plt.show()
```
